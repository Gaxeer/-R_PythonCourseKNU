# Analyzing Ukraine Car prices

This file cantains price of the car in Ukraine as to 2020 year. Prices are taken from car sales sites.

#### https://www.kaggle.com/dimakyn/vehicles-price-2020-ukraine

#### brand - the brand of the car
#### model - the model of the car
#### year - the year of creation of the model
#### body - the type of car body
#### price - the price in USD
#### mileage - the mileage of the car
#### fuel - the type of fuel the car is using
#### transmission - the type of the car transmission

```python
import pandas as pd

carsUkraine = pd.read_csv('vehicle_price(2020).csv', index_col=0)

carsUkraine.head(10)

carsUkraine.dtypes

carsUkraine["price$"] = carsUkraine["price$"].str.replace(' ', '')

carsUkraine.drop(carsUkraine.index[carsUkraine['price$'] == 'договірна'], inplace=True)

carsUkraine["price$"] = pd.to_numeric(carsUkraine["price$"])

carsUkraine = carsUkraine.dropna()

carsUkraine.head(10)
```

### Question 1: Find the most expensive car with between 2010 and 2015

```python
carsOfProperYears = carsUkraine[carsUkraine["year"].between(2010,2015)]
carsOfProperYears[carsOfProperYears["price$"] == carsOfProperYears["price$"].max()]
```

### Question 2: Find how many cars run on diesel, gas or petrol

```python
countByFuel = carsUkraine.groupby('fuel').size().reset_index()
countByFuel.rename(columns = {0:'count'}, inplace = True)
countByFuel
```

### Question 3: Find statistic of mileage for every type of transmission

```python
carsUkraine.groupby('transmission').car_mileage.agg(['max', 'median', 'mean', 'count'])
```

### Question 4: Group cars by age to Old and New

```python
carsUkraine['age'] = carsUkraine['year'].apply(lambda x : 'Old' if x < 2005 else 'New')
carsUkraine
```

### Question 5: Find the most popular model with price lower than median

```python
carsUkraine[carsUkraine['price$'] <= carsUkraine['price$'].median()].groupby('model').size().idxmax()
```
